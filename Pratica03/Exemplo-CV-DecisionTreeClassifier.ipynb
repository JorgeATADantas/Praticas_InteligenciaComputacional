Utlizando a biblioteca [pandas](https://pandas.pydata.org/docs/user_guide/index.html) para carregar os dados

```python
import pandas as pd

df_test = pd.read_csv("test.csv")
df_train = pd.read_csv("train.csv")

X_train = df_train.drop(columns=["ID_code", "target"]) #remove as colunas ID_code e target
y_train = df_train["target"] #iremos tentar prever a coluna target
```

Removendo o atributo `ID_code`, porque essa coluna não trará informação para a classificação

```python
X_test = df_test.drop(columns=["ID_code"])
X_test
```

Exemplo de código para treinar e predizer o label dos dados da comptetição do Santander

```python
#!unzip santander-customer-transaction-prediction.zip

from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_validate

classifier = DecisionTreeClassifier(random_state=0,max_depth=3) #instancia o classificador
#classifier = GaussianNB() #instancia o classificador
#classifier = KNeighborsClassifier() #instancia o classificador

scores = cross_validate(classifier, X_train, y_train, scoring=["f1_macro", 'precision'], cv=10)

scores["test_f1_macro"].mean()
```

```python
scores
```

```python

classifier.fit(X_train, y_train) #treina o classificador
predictions = classifier.predict(X_test) #prediz a classe (ou seja, a coluna target)
```

Atribuindo as predições para o dataframe com os dados de teste. Cada linha receberá a sua respectiva predição.

```python
df_test["target"] = predictions
```

```python
df_test[["ID_code","target"]] #exibe as colunas que serão enviadas para o Kaggle
```
